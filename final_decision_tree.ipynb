# Iris DecisionTree 

## Importanat Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as datasets
%matplotlib inline
```

## Loading dataset

```python
data=pd.read_csv('iris.csv')
data.head()
```

```python
data.info()
```

```python
x=data[['sepal length in cm', 'sepal width in cm', ' petal length in cm','petal width in cm']].values
x[0:5]
```

```python
y=data[['class']]
y[0:5]
```

## Split data in test/train set

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

## Build the Decision tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classtree=DecisionTreeClassifier(criterion="entropy",max_depth=4)
classtree
```

```python
classtree.fit(x_train,y_train)
```

```python
predtree=classtree.predict(x_test)
```

```python
print(predtree[0:5])
print(y_test[0:5])
```

## Accuracy of Model

```python
from sklearn import metrics
```

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test,predtree))
```
